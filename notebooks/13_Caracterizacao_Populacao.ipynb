```python
import pandas as pd
import numpy as np
import unidecode

data_d.columns = data_d.columns.str.strip().str.lower()

col_degrad = [c for c in data_d.columns if "degradacao" in c][0]

data_d[col_degrad] = (
    data_d[col_degrad]
    .astype(str)
    .str.strip()
    .str.lower()
    .apply(unidecode.unidecode)  # remove acentos
)

# Dicionário ajustado sem acentos
mapa_degradacao = {
    "nao possui degradacao": 0,
    "pouco degradado": 1,
    "degradacao media": 2,
    "muito degradado": 3
}

# Aplicar mapeamento
data_d["degradacao_ord"] = data_d[col_degrad].map(mapa_degradacao)

# Conferir valores únicos e se deu certo
print("Valores únicos de degradação bruta:", data_d[col_degrad].unique())
print("Valores únicos de degradação codificada:", data_d["degradacao_ord"].unique())
```

```python
import pandas as pd

# 1. carregar CSV
data = escasso

col = 'enfrenta_dificuldades_para_manter_atividade_produtiva_e_o_atendimento_domestico_em_virtude_de_escass'

# 2. lista das variantes que devem virar 'Sim, julho a novembro'
variants = [
    'De dezembro a maio',
    'agosto a outubro',
    'FALTA DE ENERGIA PARA DISTRIBUIÇÃO DE AGUA MECANIZADA',
    'AGOSTO OUTUBRO',
    'Setembro a novembro',
    'AGOSTO A OUTUBRO',
    'agosto e setembro',
    'setembro a novembro',
    'AGOSTO/SETEMBRO',
    'junho a dezembro',
    'QUANDO FALTA ENERGIA NA BOMBA',
    'NÃO TEM ÁGUA NA PROPRIEDADE',
    'De agosto a setembro',
    'agosto a novembro',
    'meses de agosto a dezembro',
    'setembro',
    'setembro a outubro'
]

# 3. normalizar e criar máscara (case-insensitive)
norm = data[col].astype(str).str.strip().str.lower()
variants_norm = [v.strip().lower() for v in variants]

mask_variants = norm.isin(variants_norm)

# 4. aplicar a primeira substituição para unificar variantes
data.loc[mask_variants, col] = 'Sim, julho a novembro'

# 5. garantir que entradas já iguais (com variações de case/espacos) também sejam capturadas
#    (por exemplo 'Sim, julho a novembro' com maiúsculas/minúsculas diferentes)
mask_sim_julho = data[col].astype(str).str.strip().str.lower() == 'sim, julho a novembro'
data.loc[mask_sim_julho, col] = 'Sim, julho a novembro'  # padroniza o texto

# 6. renomear 'Sim, julho a novembro' para 'Sim, período de seca'
data.loc[data[col].str.strip().str.lower() == 'sim, julho a novembro', col] = 'Sim, período de seca'

# 7. conferir resultado
print(data[col].value_counts())
```
