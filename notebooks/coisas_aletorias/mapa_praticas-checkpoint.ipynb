```python
# %% [markdown]
# # Mapa Choropleth - Práticas Sustentáveis
# Este notebook gera um mapa estático mostrando a média de práticas sustentáveis
# por município do Tocantins.

# %%
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# %% [markdown]
# ## 1. Carregar dados agregados

# %%
df_mun = pd.read_csv("C:/Users/Valentine/Artigo_Mapitos/data_clean/media_praticas_por_municipio.csv")
df_mun.head()

# %% [markdown]
# ## 2. Carregar geometria dos municípios (GeoJSON simplificado do Brasil)

# %%
url_geo = "https://raw.githubusercontent.com/tbrugz/geodata-br/master/geojson/geojs-100-mun.json"
municipios = gpd.read_file(url_geo)

# código IBGE no geojson
municipios["id"] = municipios["id"].astype(int)

# filtrar Tocantins (UF=17 no IBGE: primeiros 2 dígitos do código)
municipios_to = municipios[municipios["id"] // 100000 == 17].copy()

print("Número de municípios no TO:", len(municipios_to))

# %% [markdown]
# ## 3. Merge com dados

# %%
gdf = municipios_to.merge(df_mun, left_on="id", right_on="codigo_ibge", how="left")

print("Preview do GeoDataFrame final:")
print(gdf.head())

# %% [markdown]
# ## 4. Plotar mapa coroplético

'''
# %%
fig, ax = plt.subplots(1, 1, figsize=(8, 8))


# plotar
gdf.plot(
    column="media_praticas",   # variável numérica
    cmap="YlGnBu",             # colormap (verde-azulado)
    linewidth=0.5,
    edgecolor="0.8",
    legend=True,
    ax=ax
)

ax.set_title("Média de práticas sustentáveis por município (TO)", fontsize=14)
ax.axis("off")

plt.tight_layout()
plt.show()
'''

```
